```python
%config IPCompleter.greedy=True
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
df = pd.read_csv("oasis_longitudinal.csv")
print("Demented", df["Group"].value_counts())
```

```python
df["Group"] = df["Group"].replace(["Converted"], ["Demented"])
```

```python
print(df["Group"].value_counts())

group_counts = df["Group"].value_counts()
sns.countplot(x="Group", data=df)
```

```python
def bar_chart(data,variable):
    demented = data[data["Group"] == "Demented"][variable].value_counts()
    nondemented = data[data["Group"] == "Nondemented"][variable].value_counts()
    bar_df = pd.DataFrame([demented, nondemented], index = ["demented", "nondemented"])
    print(df.head(3))
    print(bar_df)
    bar_df.plot(kind="barh", stacked = True, figsize = (8,5))
    

bar_chart(df,"M/F")

df.hist(figsize = (10,10))
plt.show()
```

```python
print(df["SES"].isnull().value_counts())
df2 = df.dropna()
print(df2["SES"].isnull().value_counts())
df2.hist(figsize=(10,10))
plt.show()
```

```python
df2 = df2.loc[df["Visit"] == 1]
df2 = df2.reset_index(drop = True)
print(df2.head(5))
df2 = df2.drop(["Hand", "Subject ID", "MRI ID", "Visit", "CDR", "MR Delay"], axis = 1)
df2["Group"] = df2["Group"].replace(["Demented", "Nondemented"], [1,0])
df2["M/F"] = df2["M/F"].replace(["M", "F"], [1,0])
print(df2.head(5))
```

```python
print(df2.head(5))


plt.subplots(5,2, figsize=(30,30))

for indx, col in enumerate(df2.columns):
    if col == "Group_Demented" or col == "Group_Nondemented" :
        continue
    ax = plt.subplot(5,2, indx +1)
#     ax.yaxis.set_ticklabels([])
    sns.distplot(df2.loc[df2.Group == 0][col], hist=True, axlabel=False,
    hist_kws = {"histtype": "step", "color": "green", 'linewidth': 2, 'alpha':1},
    kde_kws = {'linestyle':'-', 'color':'green','label':'No Dementia'})
    sns.distplot(df2.loc[df2.Group == 1][col], hist=True, axlabel=False,
    hist_kws = {"histtype": "step", "color": "red", 'linewidth': 2, 'alpha':1},
    kde_kws = {'linestyle':'--', 'color':'red','label':'Dementia'})
    ax.set_title(col)
plt.show()
    
```
